```python
desired_seasons = ['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']

# Dictionary to store the count of wins for each team
team_wins_count = {}

# Iterate through each row in the dataset
for index, row in df.iterrows():
    # Get the season and winning team
    season = row['Season']
    winning_team = row['WinningTeam']
    
    # Check if the season is in the desired seasons list
    if season in desired_seasons:
        # Increment the count for the winning team
        if winning_team in team_wins_count:
            if season in team_wins_count[winning_team]:
                team_wins_count[winning_team][season] += 1
            else:
                team_wins_count[winning_team][season] = 1
        else:
            team_wins_count[winning_team] = {season: 1}

# Print the count of wins for each team in each season
for team, wins_per_season in team_wins_count.items():
    print(f"{team}:")
    for season, wins in wins_per_season.items():
        print(f"\t{season}: {wins} wins")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'matches' DataFrame contains the provided data
# You can load your data using pd.read_csv() or other appropriate functions

# Count the number of matches won by each team in every season
matches_won_per_season = df.groupby(['Season', 'WinningTeam']).size().unstack(fill_value=0)

# Drop rows and columns with all zeros
matches_won_per_season = matches_won_per_season.loc[(matches_won_per_season != 0).any(axis=1), (matches_won_per_season != 0).any(axis=0)]

# Plotting
plt.figure(figsize=(20, 10))

# Generate a colormap with the same number of colors as unique teams
num_teams = len(matches_won_per_season.columns)
colors = plt.cm.tab20(np.linspace(0, 1, num_teams))

# Plotting each team's matches won
for i, team in enumerate(matches_won_per_season.columns):
    plt.bar(matches_won_per_season.index, matches_won_per_season[team], bottom=matches_won_per_season.iloc[:, :i].sum(axis=1), label=team, color=colors[i])

plt.title('Number of Matches Won by Each Team in Every Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches Won')
plt.legend(title='Team', bbox_to_anchor=(1.7, 0.6), loc='upper left')
plt.xticks(rotation=90)
plt.yticks([])  # Remove y-axis tick labels

# Adding integer labels to the bars with smaller font size
for team in matches_won_per_season.columns:
    for i, value in enumerate(matches_won_per_season[team]):
        if value != 0:  # Only annotate non-zero values
            plt.text(i, matches_won_per_season.iloc[i, :matches_won_per_season.columns.get_loc(team)].sum() + value/2, f'{int(value)}', ha='center', va='center', fontsize=10, weight='bold')

plt.tight_layout()
plt.show()
```
