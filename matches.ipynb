### team wise with respect to the venue wise

```python
a={}
for i in df["team1"].unique():
    df_venue_wons_matches=df[df["winner"]==i]["venue"].value_counts().reset_index()
    df_venue_total_matches=df[(df["team1"]==i)|(df["team2"]==i)]["venue"].value_counts().reset_index()
    statdium=pd.merge(df_venue_total_matches,df_venue_wons_matches,on="venue")
    statdium["percentage"]=(statdium["count_y"]/statdium["count_x"])*100
    
    a[i]=statdium
    
```

```python
import matplotlib.pyplot as plt

# Create the cross-tabulation table with margins
cross_tab = pd.crosstab(index=df["year"], columns=df["month"], values=df["count"], aggfunc=np.sum, margins=True)

# Drop the 'All' row and column (margin), if they exist
if 'All' in cross_tab.index:
    cross_tab = cross_tab.drop('All', axis=0)
if 'All' in cross_tab.columns:
    cross_tab = cross_tab.drop('All', axis=1)

# Plot the data
plt.figure(figsize=(10, 6))
for column in cross_tab.columns:
    plt.plot(cross_tab.index, cross_tab[column], label=column)

plt.title('Monthly Counts Over Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
