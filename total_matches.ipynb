```python
import ast
import pandas as pd

# Sample string of players
players_str = "['YBK Jaiswal', 'JC Buttler', 'SV Samson', 'D Padikkal', 'SO Hetmyer', 'R Ashwin', 'R Parag', 'TA Boult', 'OC McCoy', 'M Prasidh Krishna', 'YS Chahal']"

# Convert the string to a list of players
players_list = ast.literal_eval(players_str)

# Count the occurrences of each player
player_counts = {}
for player in players_list:
    if player in player_counts:
        player_counts[player] += 1
    else:
        player_counts[player] = 1

# Create a DataFrame from the player counts
player_counts_df = pd.DataFrame(list(player_counts.items()), columns=['Player', 'Count'])

print(player_counts_df)
```

```python
import ast
from collections import Counter

# Initialize an empty Counter to store player counts
total_player_counts = Counter()

# Assuming df1 is your DataFrame containing the "Team1Players" column
for index, row in df1.iterrows():
    # Parse the string into a list of players
    players_list = ast.literal_eval(row["Team1Players"])
    # Update the Counter with the counts of players in the current row
    total_player_counts.update(players_list)

# Convert the Counter to a DataFrame
player_counts_df2 = pd.DataFrame(list(total_player_counts.items()), columns=['Player', 'Total Count'])

# Sort the DataFrame by player counts
player_counts_df2 = player_counts_df2.sort_values(by='Total Count', ascending=False)

print(player_counts_df2)
```

```python
import ast
from collections import Counter

# Initialize an empty Counter to store player counts
total_player_counts = Counter()

# Assuming df1 is your DataFrame containing the "Team1Players" column
for index, row in df1.iterrows():
    # Parse the string into a list of players
    players_list = ast.literal_eval(row["Team2Players"])
    # Update the Counter with the counts of players in the current row
    total_player_counts.update(players_list)

# Convert the Counter to a DataFrame
player_counts_df = pd.DataFrame(list(total_player_counts.items()), columns=['Player', 'Matches_played'])

# Sort the DataFrame by player counts
player_counts_df = player_counts_df.sort_values(by='Matches_played', ascending=False)

player_counts_df
```
